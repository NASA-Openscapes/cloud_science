# Modes
## Jupyter Notebook has two operation modes

### `Command`: applies to the entire scripts
### `Edit`: applies to a sigle cell

<div class="alert alert-block alert-info">
    - <b>Click on the cell below to enter the <i>Edit</i> mode.</b> Recognize it by the green color on the right
<br>
<br>
    - <b>Then press <i>Esc</i> to enter the <i>Command</i> mode.</b> Blue color on the right
</div>

## Two type of cells
### `Code` Cells: execute code
### `Markdown` Cells: show formated text

There are two ways to change the type on a cell:
- Cliking on the scroll-down menu on the top

- using the letters `y` for code and `m` for markdown on `command` mode
<br>

# To execute commands

## - `Shift-Enter` : executes cell & advance to next
## - `Control-enter` : executes cell & stay in the same cell

## Now, the keyboard shortcuts
#### In command mode:
  - `s` : save changes
  <br>
  - `a`, `b` : create cell above and below
  <br>
  - `x`, `c`, `v` : cut, copy and paste cell
  <br>
  - `z` undo last change
